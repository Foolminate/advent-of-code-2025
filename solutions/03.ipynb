## [--- Day 3: Lobby ---](https://adventofcode.com/2025/day/3)

```python
def solve() -> None:
    with open("..\\data\\03.txt") as file:
        battery_banks = [line.strip() for line in file.readlines()]
    
    charges: str = '987654321'
    total_a: int = 0
    total_b: int = 0

    for bank in battery_banks:
        start_range: str = bank[:-1]
        for joltage in charges:
            if joltage in start_range:
                pos = bank.index(joltage)
                total_a += int(joltage) * 10
                break
        
        end_range: str = bank[pos + 1:]
        for joltage in charges:
            if joltage in end_range:
                total_a += int(joltage)
                break
        
        start = 0
        window_total = 0
        for end in range(len(bank) - 11, len(bank) + 1):
            window = bank[start:end]
            for joltage in charges:
                if joltage in window:
                    start += window.index(joltage) + 1
                    window_total *= 10
                    window_total += int(joltage)
                    break
        total_b += window_total

    
    print(f"Total charge (part a): {total_a}")
    print(f"Total charge (part b): {total_b}")


solve()
```
