## [--- Day 6: Trash Compactor ---](https://adventofcode.com/2025/day/6)

```python
def solve_a() -> None:
    with open("..\\data\\06.txt") as file:
        data = [line.strip().split() for line in file.readlines()]
    
    total: int = 0
    for i, operator in enumerate(data[-1]):
        if operator == '+':
            temp = 0
            calc = lambda x, y: x + y
        else:
            temp = 1
            calc = lambda x, y: x * y

        for line in data[:-1]:
            temp = calc(temp, int(line[i]))
        
        total += temp

    print(f'Grand total (part a): {total}')


solve_a()
```

```python
def solve_b() -> None:
    with open("..\\data\\06.txt") as file:
        data = [line for line in file.readlines()]
    
    data2 = [''.join(row) for row in zip(*data[:-1])]
    data2.pop()
    operators = data[-1].strip().split()
    operators.reverse()

    total: int = 0
    for operator in operators:
        if operator == '+':
            temp = 0
            calc = lambda x, y: x + y
        else:
            temp = 1
            calc = lambda x, y: x * y
        
        while data2:
            num = data2.pop()
            if num == '    ': break
            temp = calc(temp, int(num))
            
        total += temp
        
    print(f'Grand total (part b): {total}')
        

solve_b()
```
