## [--- Day 5: Cafeteria ---](https://adventofcode.com/2025/day/5)

```python
from intervaltree import IntervalTree

def solve() -> None:
    with open("..\\data\\05.txt") as file:
        data = [line.strip() for line in file.readlines()]
        
    separator_index = data.index("")
    fresh_ranges = [(int(start), int(stop) + 1) for start, stop in (line.split("-") for line in data[:separator_index])]
    ids = list(map(int, data[separator_index + 1:]))

    count: int = 0
    tree = IntervalTree().from_tuples(fresh_ranges)
    for id in ids:
        count += 1 if tree[id] else 0

    num_fresh_ids: int = 0
    tree.merge_overlaps()
    for interval in tree:
        num_fresh_ids += interval.end - interval.begin
    
    print(f'Count of fresh IDs (part a): {count}')
    print(f'Total number of fresh IDs (part b): {num_fresh_ids}')

solve()
```
