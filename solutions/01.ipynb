## [--- Day 1: Secret Entrance ---](https://adventofcode.com/2025/day/1)

```python
def solve_a() -> None:
    with open("..\\data\\01a.txt") as file:
        instructions = [line.strip() for line in file.readlines()]

    left_right = {"L": -1, "R": 1}
    dial = 50
    sequence = [dial]
    count = 0
    for instruction in instructions:
        direction, clicks = instruction[0], int(instruction[1:])
        dial += left_right[direction] * clicks
        dial %= 100
        sequence.append(dial)
        count += 1 if dial == 0 else 0

    print(f"Number of times dial returned to origin: {count}")


solve_a()
```

```python
def solve_b() -> None:
    with open("..\\data\\01a.txt") as file:
        instructions = [line.strip() for line in file.readlines()]

    dial: int = 50
    count: int = 0

    for instruction in instructions:
        direction: str = instruction[0]
        clicks = int(instruction[1:])

        if direction == "L":
            count += ((100 - dial) % 100 + clicks) // 100
            dial = (dial - clicks) % 100
        else:
            count += (dial + clicks) // 100
            dial = (dial + clicks) % 100

    print(f"Number of times any click caused the dial to point at 0: {count}")


solve_b()
```
