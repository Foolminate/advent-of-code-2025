## [--- Day 1: Secret Entrance ---](https://adventofcode.com/2025/day/1)

```python
def solve() -> None:
    with open("..\\data\\01a.txt") as file:
        instructions = [line.strip() for line in file.readlines()]

    dial: int = 50
    count_a: int = 0
    count_b: int = 0

    for instruction in instructions:
        direction: str = instruction[0]
        clicks = int(instruction[1:])

        if direction == "L":
            count_b += ((100 - dial) % 100 + clicks) // 100
            dial = (dial - clicks) % 100
        else:
            count_b += (dial + clicks) // 100
            dial = (dial + clicks) % 100

        count_a += 1 if dial == 0 else 0

    print(f"Number of times the dial stopped on the origin: {count_a}")
    print(f"Number of times the dial touched the origin: {count_b}")


solve()
```
