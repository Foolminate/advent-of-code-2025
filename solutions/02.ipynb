## [--- Day 2: Gift Shop ---](https://adventofcode.com/2025/day/2)

```python
import re

def solve() -> None:
    with open("..\\data\\02.txt") as file:
        ranges = file.readline().split(',')

    total_a: int = 0
    total_b: int = 0
    pattern = re.compile(r'(\d+?)\1+$')

    for num_range in ranges:
        start, end = map(int, num_range.split('-'))
        for num in range(start, end + 1):
            digits = str(num)
            num_digits = len(digits)
            middle = num_digits // 2

            if digits[:middle] == digits[middle:]:
                total_a += num

            if re.match(pattern, digits):
                total_b += num
            
    print(f"Total sum of all invalid numbers: {total_a}")
    print(f"Total sum of all invalid numbers: {total_b}")


solve()
```
