## [--- Day 4: Printing Department ---](https://adventofcode.com/2025/day/4)

```python
import numpy as np
from scipy.ndimage import uniform_filter

def solve() -> None:
    with open("..\\data\\04.txt") as file:
        rows = [line.strip() for line in file.readlines()]

    grid = np.array([1 if char == "@" else 0 for row in rows for char in row]).reshape(len(rows), len(rows[0]))

    removals: list[int] = [-1]

    while removals[-1] !=0:
        grid_sum = uniform_filter(grid.astype(np.float32), size=3, mode='constant', cval=0) * 9
        grid_sum = (grid_sum.astype(int) - grid) * grid

        removals.append(np.count_nonzero((grid == 1) & (grid_sum < 4)))
        grid = grid * ((grid_sum >= 4).astype(int))

    print(f'Total rolls accessible (part a): {removals[1]}')
    print(f'Total rolls removed (part b): {sum(removals[1:-1])}')

solve()
```
